```python
# a function to plot tests images and predicted labels
def plot_prediction(cls_obj):
    _, _, test_dir = cls_obj._prepare_data()
    test_dir_list = [test_dir + repr(i) + '/' for i in range(1, 10)]
    test_img_list = [glob.glob(test_dir_list[i] + '*.JPG')[np.random.randint(4)] for i in range(len(test_dir_list))]
    fig = plt.figure(figsize=(6, 6))
    plt.axis('off')
    for i in range(len(test_dir_list)):
        ax = fig.add_subplot(3, 3, i + 1)
        label = cls_obj.predict(test_img_list[i])
        plt.imshow(plt.imread(test_img_list[i]))
        plt.xticks([])
        plt.yticks([])
        plt.title('Digit: {0}'.format(label), fontweight='bold')
    plt.tight_layout(pad=0, h_pad=0, w_pad=0)
    plt.show()
```

```python
plot_prediction(slc)
```
